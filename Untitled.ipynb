<b>Feature Selection<b>

```python
# Using Pearson Correlation
plt.figure(figsize=(20,12)) 
sns.heatmap(df.corr(), annot=True, cmap=plt.cm.CMRmap_r)
```

```python
def correlation(df, threshold):
    col_corr = set() 
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i] 
                col_corr.add(colname)
    return col_corr
```

```python
corr_features = correlation(df, 0.98)
len(set(corr_features))
```

```python
set(corr_features)
```

```python
# dropping highly correlated features
df2 = df.drop(['area_mean','area_worst', 'perimeter_mean', 'perimeter_worst'], axis=1)
df2.head()
```

```python
df2.tail()
```

<b>Feature Scaling<b>

```python
cols_to_scale = [x for x in df2.columns.values]

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df2[cols_to_scale] = scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.tail()
```

```python
X = df2.drop(['diagnosis'], axis = 1) 
X.head()
```

```python
y = df2['diagnosis']
y.head()
```

<b>Splitting the data to prevent overfitting<b>

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

<b>Model building<b>

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
y_pred = svc_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
cm = np.array(confusion_matrix(y_test, y_pred, labels=[1,0]))
confusion = pd.DataFrame(cm, index=['is_cancer', 'is_healthy'],
                         columns=['predicted_cancer','predicted_healthy'])
confusion
```

```python
sns.heatmap(confusion, annot=True)
```

```python
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```

<b>Conclusion: SVM model indeed performed better than Logistic Regression<b>
